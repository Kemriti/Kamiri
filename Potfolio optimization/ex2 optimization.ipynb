```python
import pandas as pd
import numpy as np
```

```python
c= [10,15,16,18]
```

```python
df = pd.DataFrame(c)
```

```python
df.columns=["Price"]
```

```python
df
```

```python
df["Sreturn"]=(df.Price-df.Price.shift())/(df.Price.shift())
```

```python
df["Creturn"]=np.log(df.Price/df.Price.shift())
```

```python
df
```

```python
Sreturn = np.prod(df.Sreturn)
Creturn = np.sum(df.Creturn)
```

```python
Sreturn,Creturn
```
