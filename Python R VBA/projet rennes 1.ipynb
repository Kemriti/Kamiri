```python
# URL de la page de l'Insee avec les taux d'inflation
url = 'https://www.insee.fr/fr/statistiques/4805248#graphique-figure3_radio1'

# Envoyer une requête GET pour récupérer la page web
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Lire le tableau des données en utilisant pandas
    dfs = pd.read_html(response.text)
    # Supposons que le premier tableau contient les données d'inflation
    inflation_data = dfs[0]
    
    # Vous pouvez maintenant travailler avec les données d'inflation
    print(inflation_data)
else:
    print("Échec de la requête HTTP")

# Assurez-vous de traiter les données selon vos besoins spécifiques (filtrage, nettoyage, etc.).
```

```python
str(inflation_data.at[1, "Trimestre"])[0:4]
```

```python
import pandas as pd

# Supposons que "inflation_data" est votre DataFrame

# Utilisation d'une boucle for pour parcourir les valeurs de la colonne "Trimestre"
for i, trimestre in enumerate(inflation_data["Trimestre"]):
    # Convertir la valeur du trimestre en chaîne de caractères et obtenir le dernier caractère
    dernier_caractere = str(trimestre)[-1]
    
    # Vérifier si le dernier caractère est égal à "1"
    if dernier_caractere == "1":
        # Modifier la valeur du trimestre en concaténant "1" à la fin
        inflation_data.at[i, "Trimestre"] = str(inflation_data["Trimestre"].iloc[i])[0:5] + "01"

    elif dernier_caractere == "2":
        # Modifier la valeur du trimestre en concaténant "1" à la fin
        inflation_data.at[i, "Trimestre"] = str(inflation_data["Trimestre"].iloc[i])[0:5] + "02"
    elif dernier_caractere == "3":
        # Modifier la valeur du trimestre en concaténant "1" à la fin
        inflation_data.at[i, "Trimestre"] = str(inflation_data["Trimestre"].iloc[i])[0:5] + "03"
    elif dernier_caractere == "4":
        # Modifier la valeur du trimestre en concaténant "1" à la fin
        inflation_data.at[i, "Trimestre"] = str(inflation_data["Trimestre"].iloc[i])[0:5] + "04"


```
